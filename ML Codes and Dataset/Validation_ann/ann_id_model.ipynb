### Neural Network 

```python
#Importing Libraries for data preparation
import pandas as pd
import numpy as np
#Read Necessary files
dataset = pd.read_csv("datasetall.csv")
```

```python
dataset = np.asmatrix(dataset)

Y_data = dataset[:,4]
X_data = dataset[:,:-1]
```

```python
X_data_norma1 = dataset[:,3]
X_data_max = X_data_norma1.max()
X_data_norma1 = X_data_norma1/X_data_max
X_data[:,3] = X_data_norma1
Y_max = Y_data.max()
Y_data = Y_data/Y_data.max()
```

```python
X_data = pd.DataFrame(X_data)
Y_data = pd.DataFrame(Y_data)
```

Split in training and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.10, random_state=42)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

ANN model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Define model
model = keras.Sequential()
model.add(layers.Dense(16, input_dim=4, activation= "sigmoid"))
model.add(layers.Dense(8, activation= "sigmoid"))
model.add(layers.Dense(1))
model.summary() #Print model Summary
```

```python
# Compile model
model.compile(loss= "mean_squared_error" , optimizer="adam", metrics=["mean_squared_error"])
```

```python
model.fit(X_train, Y_train, epochs=500)
```

Inference

```python
from sklearn.metrics import mean_squared_error
pred= model.predict(X_test)
score = np.sqrt(mean_squared_error(Y_test,pred))
print (score)
```

```python
model.save("id_pred.h5")
```

```python
print(pred*Y_max)
print(Y_test*Y_max)
```

```python
import pandas as pd
import numpy as np
a=pred*Y_max
b=Y_test*Y_max
predicted=np.hstack((a,b))
pd.DataFrame(predicted).to_csv('output.csv')
```

### Random Forest Regressor

```python
from tensorflow.python.ops.gen_array_ops import reshape
import random
from sklearn.ensemble import RandomForestRegressor
random.seed(42)

rf = RandomForestRegressor(n_estimators=10)
rf.fit(X_train, Y_train.values.ravel())
```

```python
predrf=rf.predict(X_test)
score = np.sqrt(mean_squared_error(Y_test,predrf))
print (score)
```
